```python
# Example for S&P 500
sp_500_url = "https://www.slickcharts.com/sp500"
response = requests.get(sp_500_url)
soup = BeautifulSoup(response.content, "html.parser")

# Find the first company in the table
table = soup.find("table")
first_row = table.find("tbody").find("tr")
columns = first_row.find_all("td")
company_name = columns[1].text  # Second column: Company Name
symbol = columns[2].text  # Third column: Symbol

print(f"First S&P 500 company: {company_name} ({symbol})")
```
